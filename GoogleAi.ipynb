```python
import google.generativeai as genai
```

```python
f= open('/Users/vijaykoneri/iCloud Drive (Archive) - 1/Desktop/AI:ML/Gen Ai/googleAiApikey.txt')
key=f.read()

genai.configure(api_key=key)
```

```python
for m in genai.list_models():
    print(m.name)
```

```python
model=genai.GenerativeModel(model_name='gemini-1.5-flash')

prompt= ''' give me a list of allthe models one shall study in data science, names are enough'''

response=model.generate_content(prompt)
print(response.text)
```

```python
conFig= genai.types.GenerationConfig(temperature=0.1, top_p=.1, top_k=32)

usrPmpt= '''what is feature selection in data science? explain in detail'''

res= model.generate_content(usrPmpt,generation_config=conFig)

print(res.text)
```

```python
conFig= genai.types.GenerationConfig(temperature=0.9, top_p=.1, top_k=32)

usrPmpt= '''what is feature selection in data science? explain in detail'''

res= model.generate_content(usrPmpt,generation_config=conFig)

print(res.text)
```
